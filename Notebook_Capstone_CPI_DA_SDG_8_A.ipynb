```python
# Import and Show the Wage Data for Each Province in 2016-2022
url1 = 'https://raw.githubusercontent.com/JCalvinL19/CapstoneDAGigih/main/Data_Gaji_Provinsi.csv'
df1 = pd.read_csv(url1, sep =";")
df1.head()
```

```python
# Counting the Number of Provinces
df1[df1['Provinsi'] != 'Indonesia']['Provinsi'].nunique()
```

```python
df1[df1['Provinsi'] != 'Indonesia']['Tahun'].value_counts()
```

> [Provinsi] means the province that include 34 provinces in Indonesia, [Tahun] means the year, and [Nilai] indicates the average wage of the residents of the respective province



```python
# Replacing the Comma into Dot for Each Values in Nilai Column
df1['Nilai'] = df1['Nilai'].str.replace(',', '.').astype(float)
df1['Nilai'].head()
```

```python
# Showing some Informations about the Dataframe
df1.info()
```

```python
# Showing the Average Wage for All Provinces in 2016-2022
df1[df1['Provinsi'] == 'Indonesia']
```

```python
# Import and Show the Provincial Minimum Wages Data in 2016-2022
url_ump = 'https://raw.githubusercontent.com/JCalvinL19/CapstoneDAGigih/main/Data_UMP.csv'
df_ump = pd.read_csv(url_ump, sep =";")
df_ump.head()
```

```python
# Show Some Informations about the Dataframe
df_ump.info()
```

```python
df_ump['Provinsi'] = df_ump['Provinsi'].str.title()
df_ump.head()
df_ump[df_ump['Provinsi'] == 'Indonesia']
```

```python
df1[df1['Tahun'] == 2022].head(35)
```

```python
df_ump['Provinsi'] = df_ump['Provinsi'].replace(['Di Yogyakarta', 'Dki Jakarta'], ['DI Yogyakarta', 'DKI Jakarta'])
df_ump[df_ump['Tahun'] == 2022]
```

```python
# Merge the Provincial Minimum Wage Data with the Average Wage Data in 2022
merged_data = pd.merge(df_ump, df1, on=['Provinsi', 'Tahun'])
merged_data = merged_data[merged_data['Tahun'] == 2022]
merged_data['Disparity'] = merged_data['Nilai'] - merged_data['UMR/UMP']
merged_data['Percentage of Disparity (%)'] = (merged_data['Disparity'] * 100) / merged_data['UMR/UMP']
merged_data
```

```python
# Show the average of the wage and provincial minimum wage
wage_indonesia_2022 = df1[(df1['Provinsi'] == 'Indonesia') & (df1['Tahun'] == 2022)]['Nilai'].values
umr_indonesia_2022 = df_ump[(df_ump['Provinsi'] == 'Indonesia') & (df_ump['Tahun'] == 2022)]['UMR/UMP'].values
print("The average wage of all Indonesians in 2022 is", np.mean(wage_indonesia_2022))
print("The average provincial minimum wage in 2022 is", np.mean(umr_indonesia_2022))
```

```python
# Filter the DataFrame where the wage is below UMP
provinces_below_minimum_wage = merged_data[(merged_data['Disparity'] < 0) & (merged_data['Provinsi'] != 'Indonesia')]

# Count the number of provinces where wage is less than minimum wage
count = len(provinces_below_minimum_wage)

# Print the count and list of provinces
print(f"Number of Provinces where Its Residents' Average Wage is Lower Than the Provincial Minimum Wage: {count}")
provinces_below_minimum_wage[['Provinsi', 'Disparity', 'Percentage of Disparity (%)']].sort_values(by = 'Percentage of Disparity (%)', ascending = False).reset_index(drop=True)
```

```python
# Filter the DataFrame where the wage is above UMP
provinces_above_minimum_wage = merged_data[(merged_data['Disparity'] > 0) & (merged_data['Provinsi'] != 'Indonesia')]

# Count the number of provinces where wage is less than minimum wage
count = len(provinces_above_minimum_wage)

# Print the count and list of provinces
print(f"Number of Provinces where Its Residents' Average Wage is Higher Than the Provincial Minimum Wage: {count}")
provinces_above_minimum_wage[['Provinsi', 'Disparity', 'Percentage of Disparity (%)']].sort_values(by = 'Percentage of Disparity (%)', ascending = False).reset_index(drop=True)
```

```python
url_ea = "https://raw.githubusercontent.com/raflisihite/CapstoneDataset/main/EAR-Eco-Sex.csv"
df9 = pd.read_csv(url_ea)
df9.head()
```

```python
#Calculate the diff value of earning to determine the wage gap on each sector
dmerge['diff'] = abs(dmerge.value_x - dmerge.value_y)
dmerge['maxx'] = dmerge[['value_x', 'value_y']].values.max(1)
dmerge['percent'] = (dmerge['diff'] / dmerge['maxx']) * 100
dmerge[['economicactivity','time','diff','percent']].sort_values(by=['economicactivity','time','percent',], ascending=True).head()
```

```python
# Average wage difference on each sector in percent
dmerge.groupby(['economicactivity'])['percent'].mean().reset_index().sort_values(by=['percent'], ascending=False)
```

# Results of the Study

The analysis of wage disparities in Indonesia reveals several key findings. Firstly, in the majority of provinces, workers earn wages above the Provincial Minimum Wage (UMP), indicating that many regions have met or exceeded government-set wage standards. However, there are 11 provinces found in Sumatra, Kalimantan, and Sulawesi, where wages still fall below the UMP, highlighting regional disparities.

Gender-based wage gaps persist throughout the period from 2016 to 2022. On average, men earn approximately 26.21% more than women, with women's incomes consistently around 20.75% lower. Even on the same education level, females earn less than males, with 25-39% wage gap on each education level. Similarly, individuals with disabilities face significant wage disparities, with non-disabled workers earning approximately 47.45% more than their disabled counterparts. This stark inequality underscores the need for continued efforts to address gender and disability status disparities in wage.

The sector-wise analysis reveals that in sectors such as Agriculture and Construction, wages consistently fall below the UMP, suggesting that these industries may not provide sufficient income for workers. On the other hand, sectors like Manufacturing, Public Administration, Mining and Quarrying, Trade and Transportations generally offer wages above the minimum wage. Notably, the Construction sector stands out as a sector where women earn more than men, contrary to expectations. This indicates a need for further examination of wage dynamics within specific industries.

In summary, while many provinces in Indonesia meet or exceed minimum wage standards, disparities persist based on gender, disability status, and economic sector. Efforts to address these disparities are crucial to ensuring fair and adequate wages for all workers in the country.
